# Image Augumentation


```python
# import required lib 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Creating augmentation on training variable

train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
```

```python
# creating augmentation on testing variable

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# passing training data to train variable
xtrain = train_datagen.flow_from_directory('/content/drive/MyDrive/Flowers-Dataset/flowers',
                                           target_size=(64,64),
                                           class_mode='categorical',
                                           batch_size=100)
```

#  Creating model and adding layers

```python
# Importing required lib

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
```

```python
model = Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
```

# Hidden Layer

```python
model.add(Dense(300,activation='relu'))
model.add(Dense(150,activation='relu'))
```

# Output Layer

```python
model.add(Dense(4,activation='softmax'))
```

# Compile the model

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```
